```python
import rasterio
import matplotlib.pyplot as plt

# Read the .tif file
file_path = '/Data_large/marine/PythonProjects/SAR/sarpyx/focused_data/out.tif'  # Replace with your .tif file path
with rasterio.open(file_path) as src:
    tif_data = src.read(1)  # Read the first band
    profile = src.profile

print('TIF shapes: ', tif_data.shape)
# DATA TYPE
print('Data type: ', tif_data.dtype)



```

```python
# Calculate statistical measures for better contrast
mean_val = tif_data.mean()
std_val = tif_data.std()

# Set vmin and vmax based on statistical measures (2 standard deviations)
vmin = max(tif_data.min(), mean_val - 1.5 * std_val)
vmax = min(tif_data.max(), mean_val + 1.5 * std_val)

print(f'Mean: {mean_val:.6f}')
print(f'Std: {std_val:.6f}')
print(f'Calculated vmin: {vmin:.6f}')
print(f'Calculated vmax: {vmax:.6f}')
```

```python
# Plot the .tif file
plt.figure(figsize=(62, 62), dpi=400)
plt.imshow(tif_data[:,:], cmap='gray', vmin=vmin, vmax=vmax)  # Adjust the slice as needed
plt.title('TIF File Visualization')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.savefig('1.png')
plt.show()
plt.close()
```

```python
import rasterio
import matplotlib.pyplot as plt

# Read the .tif file
file_path = '/Data_large/marine/PythonProjects/SAR/sarpyx/focused_data/range_compressed/s1a-s1-raw-s-vh-20150107t054312-20150107t054337-004061-004e74_S1_RC.tif'  # Replace with your .tif file path
with rasterio.open(file_path) as src:
    tif_data = src.read(1)  # Read the first band
    profile = src.profile


# Calculate statistical measures for better contrast
mean_val = tif_data.mean()
std_val = tif_data.std()

# Set vmin and vmax based on statistical measures (2 standard deviations)
vmin = max(tif_data.min(), mean_val - 2 * std_val)
vmax = min(tif_data.max(), mean_val + 2 * std_val)

print(f'Mean: {mean_val:.6f}')
print(f'Std: {std_val:.6f}')
print(f'Calculated vmin: {vmin:.6f}')
print(f'Calculated vmax: {vmax:.6f}')


# Plot the .tif file
plt.figure(figsize=(12, 12))
plt.imshow(tif_data[:], cmap='viridis', vmin=vmin, vmax=vmax)  # Adjust the slice as needed
plt.colorbar(label='Pixel Values')
plt.title('TIF File Visualization')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
