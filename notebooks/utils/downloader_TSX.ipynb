# CCM Collection Data Downloader Tutorial

This notebook demonstrates how to search and download data from the CCM (Commercial Constellation Mission) collection using the Copernicus Data Space.

## Available Platform Short Names

When querying the CCM collection, you can filter by specific SAR missions using the `platformShortName` attribute. Here are the available options:

| Platform Code | Full Mission Name | Frequency Band | Typical Polarisation Modes |
|----------------|-------------------|----------------|-----------------------------|
| `CS00` | COSMO-SkyMed Constellation | X-band (9.6 GHz) | Single (HH or VV), Dual (HH/VV or HH/HV), Quad (HH/HV/VH/VV, 2nd Gen) |
| `CS01` | COSMO-SkyMed 1 | X-band (9.6 GHz) | Single (HH or VV), Dual (HH/VV) |
| `CS02` | COSMO-SkyMed 2 | X-band (9.6 GHz) | Single (HH or VV), Dual (HH/VV) |
| `CS03` | COSMO-SkyMed 3 | X-band (9.6 GHz) | Single (HH or VV), Dual (HH/VV) |
| `CS04` | COSMO-SkyMed 4 | X-band (9.6 GHz) | Single (HH or VV), Dual (HH/VV) |
| `IE00` | ICEYE Constellation | X-band (9.65 GHz) | Single (VV or HH depending on satellite), some Dual (VV/HH) |
| `PAZ1` | PAZ (HISDESAT) | X-band (9.65 GHz) | Single (HH or VV), Dual (HH/VV), compatible with TerraSAR-X modes |
| `RS02` | RADARSAT-2 | C-band (5.405 GHz) | Single (HH, VV, HV, VH), Dual, Quad-pol (HH/HV/VH/VV) |
| `TX01` | TerraSAR-X | X-band (9.65 GHz) | Single (HH or VV), Dual (HH/VV, HH/HV), Quad-pol (experimental) |

### Usage Example

To query a specific platform, use the `attributes` parameter:

```python
searcher.query_by_filter(
    collection_name='CCM',
    attributes={'platformShortName': 'TX01'}  # For TerraSAR-X
)
```

You can also query the entire constellation (e.g., `CS00` for all COSMO-SkyMed satellites) or individual satellites (e.g., `CS01`, `CS02`, etc.).

````python
from phidown.search import CopernicusDataSearcher
# You can modify these parameters as needed

searcher = CopernicusDataSearcher()
# Configure the search parameters
searcher.query_by_filter(
    collection_name='CCM',
    product_type=None,
    orbit_direction=None,
    cloud_cover_threshold=None,
    aoi_wkt=None,  # Use the defined AOI
    start_date='2014-05-03T00:00:00',
    end_date='2025-10-10T04:00:00',
    top=1000,
    attributes={'platformShortName':'IE00'}
)

df = searcher.execute_query()
print(f"Number of results: {len(df)}")
# Display the first few rows of the DataFrame
searcher.display_results(top_n=150)
````
