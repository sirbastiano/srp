```python
from sarpyx.snap.engine import GPT

def process_sar_data(product_path: str, outdir: str) -> None:
    """Process SAR data with default parameters.
    
    Args:
        product_path (str): Path to the SAR product (.SAFE directory)
        outdir (str): Output directory for processed data
    """
    tool = GPT(product=product_path, outdir=outdir, format='BEAM-DIMAP', gpt_path=None)
    tool.Calibration()
    tool.AdaptiveThresholding()
    tool.ObjectDiscrimination(min_target_m=5, max_target_m=500) # type: ignore
    print(f'Processing completed for {product_path}')


# Example usage
process_sar_data('/Data_large/marine/PythonProjects/SAR/sarpyx/S1C_IW_GRDH_1SDV_20250602T062629_20250602T062654_002600_005645_FD4F_COG.SAFE', 
                 '/Data_large/marine/PythonProjects/SAR/sarpyx/data/tmp3')
```
